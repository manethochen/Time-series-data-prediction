## 3.4 Model fitting and residual diagnostics

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model =SARIMAX(train_log,
                                   order=(2,1,1),
                                   seasonal_order=(1,1,1,7),
                                    enforce_stationarity=False,
                                    enforce_invertibility=False)
model_fitted =model.fit(diso=-1)
residuals_SARIMA = pd.DataFrame(model_fitted.resid)[9:]
```

```python
print(model_fitted.summary())
```

use fit function to build function

in roder to predict ,we will loss previous 9 datas

## 3.5 residual diagnostics

```python
plt.figure()
plt.plot(pd.DataFrame(residuals_SARIMA))
plt.title('SAMIRA(2,1,1),(0,1,1,7) MSR:%.4f' % np.mean((residuals_SARIMA.values)**2))
plt.show()
```

```python
plt.figure()
smt.graphics.tsa.plot_acf(residuals_SARIMA,lags=30,alpha =0.05)
smt.graphics.tsa.plot_pacf(residuals_SARIMA,lags=30,alpha=0.05)
plt.show()
```

It decrease dramatically and exsist in the blue areas so we can conlude it is stationary

## 3.6 Prediction

```python
plt.figure()
plt.plot(train,label ='train set')
plt.plot(validation.index,pd.DataFrame(forecast_sarima), label='SARIMA')
plt.plot(validation,label='validation set')
plt.legend()
plt.title('SARIMA(Dynamic Forecast)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# 5. Drift Method

The drift method defines the forecast for the time point T + h as


```python
Y_1 = train.values[0]
Y_T = train.values[-1]


def drift_naive():
    drift = []
    for i in range(1,16):
        drift.append(Y_T + i*(Y_T-Y_1)/189)
    return drift

drift_pred = drift_naive()    
```

```python
mse_drift = MSE(drift_pred ,validation.values)
testerror_drift = MSE(drift_pred,validation.values)/15

print('Seasonal Naive MSE : ' , mse_drift)
print('Test error : ' , testerror_drift)
```

# 6. Holt-Winters smoothing
Holt-Winters smoothing is an Exponential smoothing method for data with seasonality.
Additive seasonality and Multiplicative seasonality.
a) Additive seasonality: the seasonal variation is constant along the
trend
b) Multiplicative seasonality: the seasonal variation is changing along
the trend

## 6.1 Additive Holt-Winters smoothing


The forecast is 


```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
fit_add = ExponentialSmoothing(train,seasonal_periods=7, trend='add', seasonal='add').fit() # Holt-Winters additive smoothing
fit_mul = ExponentialSmoothing(train, seasonal_periods=7, trend='add', seasonal='mul').fit() # Holt-Winters multiplicative smoothing
```

```python
results = pd.DataFrame(index=[r"$\alpha$",\
                              r"$\beta$",\
                              r"$\gamma$",\
                              r"$l_0$",\
                              "$b_0$",\
                              "SSE"])
# ExponentialSmoothing() object has following attributes
params = ['smoothing_level', \
          'smoothing_slope', \
          'smoothing_seasonal', \
          'initial_level', \
          'initial_slope']

# check out the performance of additive and multiplicative
results["Additive"]       = [fit_add.params[p] for p in params] + [fit_add.sse]
results["Multiplicative"] = [fit_mul.params[p] for p in params] + [fit_mul.sse]
print(results)
```

```python
# Obtain fitted values from the fittedvalues attribute
smooth_add = fit_add.fittedvalues
smooth_mul = fit_mul.fittedvalues

# Plot all together
plt.figure(figsize=(15,6))
plt.plot(train, 'b-', label = 'Original data')
plt.plot(smooth_add, 'r-.',label = 'Holt-Winters additivie')
plt.plot(smooth_mul, 'g--', label = 'Holt-Winters multiplicative')
plt.xlabel("Time")
plt.ylabel("Daily Cases")
plt.title("Holt-Winters' methods")
plt.legend()
plt.show()
```

```python
# Forecast 12 more data points
y_forecast_add = fit_add.forecast(15)
y_forecast_mul = fit_mul.forecast(15)

# And plot al together
plt.figure(figsize=(16,5))
plt.plot(validation, 'b-',label = 'Target')
plt.plot(y_forecast_add, 'r-.',label = 'Holt-Winters additive')
plt.plot(y_forecast_mul, 'g--', label = 'Holt-Winters multiplicative')
plt.xlabel("Year")
plt.ylabel("Daily Cases")
plt.title(" Forecasts from Holt-Winters'methods")
plt.legend()
plt.show()
```

```python
mse_hw_add = MSE(y_forecast_add.values,validation.values)
testerror_hw_add = MSE(y_forecast_add.values,validation.values)/15

print('Addictive Holtwinters forecast MSE : ' , mse_hw_add)
print('Test error : ' , testerror_hw_add)
```

```python
mse_hw_mul = MSE(y_forecast_mul.values,validation.values)
testerror_hw_mul = MSE(y_forecast_mul.values,validation.values)/15

print('Multiplicative Holtwinters forecast MSE : ' , mse_hw_mul)
print('Test error : ' , testerror_hw_mul)
```
